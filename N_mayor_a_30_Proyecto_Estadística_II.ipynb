**Problema:**

Se dice que la velocidad promedio de conexión a internet en una ciudad es de 50 Mbps. Una empresa realiza una auditoría de calidad y toma una muestra aleatoria de 64 hogares, encontrando una velocidad media de 48.9 Mbps, con una desviación estándar poblacional conocida de σ = 3.2 Mbps. ¿Existe suficiente evidencia para decir que la velocidad promedio es diferente a 50 Mbps? Usa un nivel de significancia de 0.05.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Hipotesis Nula
H0 = 50

# Nivel de Significancia
a = 0.05

# Z critico
Zmenor = -1.96
Zmayor = 1.96

# Media Muestral
X = 48.9

# Varianza
o = 3.2

# Poblacion
n = 64

# Media poblacional
xn = 50

```

Aplicamos la siguiente ecuación:

image.png

```python
# Estadistico de Prueba Z
Z = (X - xn) / (o / np.sqrt(n))

# Se imprime el resultado
Z
```

```python
# Regla de Decision

if(Z < Zmenor or Z > Zmayor):
    print("Se rechaza la hipotesis nula")
else:
    print("No se rechaza la hipotesis nula")
```

```python
# Se calcula el valor p
p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

# Se crea una trama estadistica
x_axis = np.linspace(-4, 4, 100)
plt.plot(x_axis, stats.norm.pdf(x_axis, 0, 1), color='skyblue', label='Distribución Normal Estándar')

# Se agregan la regiones críticas
plt.fill_between(x_axis, stats.norm.pdf(x_axis, 0, 1), where=(x_axis < Zmenor) | (x_axis > Zmayor), color='salmon', alpha=0.5, label='Región de Rechazo')

# Se agrega el estadístico de prueba
plt.axvline(Z, color='red', linestyle='--', label=f'Estadístico de Prueba Z = {Z:.2f}')

# Se agregan titulo y etiquetas
plt.title('Prueba de Hipótesis')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(f"Estadístico de Prueba (Z): {Z:.2f}")
print(f"Valores Críticos: [{Zmenor}, {Zmayor}]")
print(f"Valor P: {p_value:.4f}")
```

**Conclusión:**

Como Z = -2.75 cae en la región de rechazo, se rechaza H₀. Hay suficiente evidencia estadística para concluir que la velocidad promedio de internet es diferente a 50 Mbps.
