**Problema:**

Se afirma que los estudiantes de ingeniería en sistemas tardan en promedio 60 segundos en resolver un cubo Rubik. Se toma una muestra aleatoria de 16 estudiantes, con un tiempo promedio de 56.8 segundos y desviación estándar muestral s = 5.1 segundos.

¿Se puede concluir que los estudiantes resuelven el cubo más rápido que el promedio nacional? Usa un nivel de significancia de 0.05.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Poblacion
n = 16

# Media Muestral
X = 56.8

# Media poblacional
xn = 60

# Varianza
s = 5.1

# Hipotesis
H = 60

# Nivel de Significancia
a = 0.05

# T critico

Tc = stats.t.ppf(a, (n-1))
```

Aplicamos la siguiente ecuación:

image.png

```python
# Estadístico de prueba T
T = (X-xn)/(s/np.sqrt(n))

# Se imprime el resultado
T
```

```python
# Regla de Decisión

if(T < Tc):
    print("Se rechaza la hipótesis nula")
else:
    print("No se rechaza la hipótesis nula.")
```

```python
# Se crea una trama estadistica
x = np.linspace(-5, 5, 1000)
y = stats.t.pdf(x, df=n-1)
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y)

# Se agregan líneas verticales para T crítico y estadístico de T
plt.axvline(x=Tc, color='red', linestyle='--', label=f'T crítico ({Tc:.2f})')
plt.axvline(x=T, color='green', linestyle='-', label=f'Estadístico de T ({T:.2f})')

# Sombra de Región de rechazo
x_rejection = np.linspace(-5, Tc, 100)
y_rejection = stats.t.pdf(x_rejection, df=n-1)
plt.fill_between(x_rejection, y_rejection, color='red', alpha=0.2, label='Región de Rechazo')

plt.title('Distribución de T de Student y Prueba de Hipótesis')
plt.xlabel('Valor T')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

**Conclusión:**

Se rechaza H₀, lo cual indica que hay suficiente evidencia para decir que los estudiantes resuelven el cubo Rubik más rápido que el promedio nacional de 60 segundos.
